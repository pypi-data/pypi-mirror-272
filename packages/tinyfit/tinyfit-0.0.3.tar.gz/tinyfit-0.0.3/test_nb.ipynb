```python
import src.tinyfit.linreg as lr
import src.tinyfit.scaling as scaling
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 200
x_values = np.random.normal(100,20,n)
rand = np.random.normal(0,5,n)
y_values = 10 + x_values + rand


x_values = scaling.minmax(x_values)
#x_values = (x_values - x_values.mean())/(x_values.max() - x_values.min())
```

```python
plt.scatter(x_values, y_values)
plt.show()
```

```python
lreg = lr.LinearRegression(lr=0.01)
```

```python
lreg.fit(x_values, y_values, iter=10000)
```

```python
lreg.W.item(), lreg.b.item()
```

```python
ls = np.linspace(x_values.min(),x_values.max(),100)
pred = ls*lreg.W.item() + lreg.b.item()
plt.plot(ls, pred)
plt.scatter(x_values, y_values, c='r', marker='x')
plt.show()
```
